```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data = np.loadtxt('ps1_data1.csv', delimiter=',')
data
```

```python
def setVariables(data_array) :

    global x
    global y
    global m
    
    x = data_array[:,0]
    y = data_array[:,1]
    m = len(data_array)
    return x, y, m
#setVariables(data)
```

```python
setVariables(data)
```

```python
plt.scatter(x, y, label="ps1_data1")
plt.title('2017 World Happiness by Country')
plt.xlabel(r'Per Capita GPD')
plt.ylabel(r'Happiness Score')
plt.legend()
```

```python
# initialize data array
X = np.ones((m, 2))
X[:,1] = data[:,0]
# initalize model
theta = np.zeros(2)
```

```python
def costJ(X, y, theta):

    dotprod = np.dot(X, theta)
    
    error = dotprod - y
    scores = np.power(error, 2)
    #scores = np.power(np.dot(x, theta)- y, 2)
    
    ans = sum(scores)/ (2*m)
    
    return ans
costJ(X, y, theta)
```

```python
def batchGradientDescent(X, y, theta, alpha, num_iterations):

    m= len(y)
    J_history = []

    theta_new = theta.copy()
    
    for ii in range(num_iterations):
        h = np.dot(X, theta_new)
        error = (h - y)
        grad= (np.matmul(X.T , error))/m
        theta_new -=  alpha*grad
        J_history.append(costJ(X, y, theta_new))
    
    return theta_new, J_history

alpha=0.01

num_iterations=1000

batchGradientDescent(X, y, theta, alpha, num_iterations)
```

```python
alpha = 0.2
iterations = 1100
```

```python
theta_new, J_history = batchGradientDescent(X, y, theta, alpha, iterations)
theta_new
```

```python
plt.plot(J_history)
plt.xlabel(r'Iteration number' )
plt.ylabel(r'$J(\theta)$')
```

```python
alphas = [0.2, 0.1, 0.01, 0.001, 0.0001]
iters = 20
for aa in alphas:
    theta_new, J_history = batchGradientDescent(X, y, theta, aa, iters)
    plt.plot(J_history, label='learning rate={0}'.format(aa) )

plt.ylabel(r'$J(\theta)$')
plt.xlabel(r'Number of Iterations' )
plt.legend(loc='best', frameon=False)
```
